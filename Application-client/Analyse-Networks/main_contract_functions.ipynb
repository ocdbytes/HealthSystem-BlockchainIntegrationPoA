```python
import matplotlib.pyplot as plt
import csv
```

```python
# ETH
# ---

x_axis_MainContractInitialisation = []
y_axis_MainContractInitialisation = []

x_axis_GetSelfKeys = []
y_axis_GetSelfKeys = []

x_axis_CheckIfPatientExists = []
y_axis_CheckIfPatientExists = []

x_axis_AddPatient = []
y_axis_AddPatient = []

# POLYGON
# -------

x_axis_MainContractInitialisation_pol = []
y_axis_MainContractInitialisation_pol = []

x_axis_GetSelfKeys_pol = []
y_axis_GetSelfKeys_pol = []

x_axis_CheckIfPatientExists_pol = []
y_axis_CheckIfPatientExists_pol = []

x_axis_AddPatient_pol = []
y_axis_AddPatient_pol = []
```

```python
with open("./main_ethereum.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')

    for dataline in data:
        if(dataline[0] == "MainContractInitialisation&Connection"):
            x_axis_MainContractInitialisation.append(dataline[3])
            y_axis_MainContractInitialisation.append(float(dataline[1]))
        elif(dataline[0] == "GetSelfKeys"):
            x_axis_GetSelfKeys.append(dataline[3])
            y_axis_GetSelfKeys.append(float(dataline[1]))
        elif(dataline[0] == "CheckIfPatientExists"):
            x_axis_CheckIfPatientExists.append(dataline[3])
            y_axis_CheckIfPatientExists.append(float(dataline[1]))
        elif(dataline[0] == "AddPatient"):
            x_axis_AddPatient.append(dataline[3])
            y_axis_AddPatient.append(float(dataline[1]))

with open("./main_polygon.csv", 'r') as f:
    data = csv.reader(f, delimiter=',')

    for dataline in data:
        if(dataline[0] == "MainContractInitialisation&Connection"):
            x_axis_MainContractInitialisation_pol.append(dataline[3])
            y_axis_MainContractInitialisation_pol.append(
                float(dataline[1]))
        elif(dataline[0] == "GetSelfKeys"):
            x_axis_GetSelfKeys_pol.append(dataline[3])
            y_axis_GetSelfKeys_pol.append(float(dataline[1]))
        elif(dataline[0] == "CheckIfPatientExists"):
            x_axis_CheckIfPatientExists_pol.append(dataline[3])
            y_axis_CheckIfPatientExists_pol.append(float(dataline[1]))
        elif(dataline[0] == "AddPatient"):
            x_axis_AddPatient_pol.append(dataline[3])
            y_axis_AddPatient_pol.append(float(dataline[1]))
```

```python
# MainContractInitialisation
plt.plot(x_axis_MainContractInitialisation, y_axis_MainContractInitialisation,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_MainContractInitialisation_pol, y_axis_MainContractInitialisation_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Main Contract Initialisation (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# GetSelfKeys
plt.plot(x_axis_GetSelfKeys, y_axis_GetSelfKeys,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_GetSelfKeys_pol, y_axis_GetSelfKeys_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Get Self Keys (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# CheckIfPatientExists
plt.plot(x_axis_CheckIfPatientExists, y_axis_CheckIfPatientExists,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_CheckIfPatientExists_pol, y_axis_CheckIfPatientExists_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Check If Patient Exists (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```

```python
# AddPatient
plt.plot(x_axis_AddPatient, y_axis_AddPatient,
         color="r", linestyle="solid", marker='o', label="ETH")
plt.plot(x_axis_AddPatient_pol, y_axis_AddPatient_pol,
         color="b", linestyle="solid", marker='o', label="POLYGON")
plt.ylabel('Time Taken (s)')
plt.xlabel('Count (freq)')
plt.title("Add Patient (ETH/POLYGON)", fontsize=20)
plt.grid()
plt.legend()
plt.show()
```
